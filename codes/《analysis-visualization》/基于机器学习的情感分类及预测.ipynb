```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import warnings
```

```python
# 加载总体结果
results = pd.read_csv('./storage-总体结果.csv',encoding="utf-8")
```

```python
comment = results['word']
list_word = [] #整合

for i in comment:
    list_word.append(i)
print(list_word)
```

# TF-IDF 特征抽取

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()  # 调用TF-TDF分词算法，将文本转化为频率
data1 = tf.fit_transform(comment.values.astype('U'))  # 先拟合，再转换，得出一个稀疏矩阵
print(data1)
```

```python
#对酒店的评分
#低于4分的为差评 否则为好评
bq = []
for i in results['评分']:  
     if float(i) <= 4:
         bq.append(0)
     else:
         bq.append(1)
```

# 划分数据集

```python
x = data1
y = bq
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# 对数据进行标准化处理
from sklearn.preprocessing import StandardScaler
# 标准化后，有的值为正，有的值为负
x_train = StandardScaler(with_mean=False).fit_transform(x_train)
x_test = StandardScaler(with_mean=False).fit_transform(x_test)
```

# 情感二分类及预测

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
#朴素贝叶斯分类算法
```

```python
from sklearn.naive_bayes import BernoulliNB
# Fitting a Naive Bayes model with default parameters
clf = BernoulliNB()
clf.fit(x_train, y_train)
```

```python
# Prediction & Evaluation
y_hat_test = clf.predict(x_test)
# Naive Bayes score
#precision_score(y_train, y_train_pred, average='micro')
print("Naive Bayes score for test set:")
print("Precision: {:.3f}".format(precision_score(y_test, y_hat_test, average='micro')), "Recall: {:.3f}".format(recall_score(y_test, y_hat_test, average='micro')))
print("F1 score: {:.3f}".format(f1_score(y_test, y_hat_test, average='micro')))
```

```python
#决策树模型
```

```python
from sklearn.tree import DecisionTreeClassifier
# Fitting a decision tree model with default parameters
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
# Prediction & Evaluation
y_hat_test = dt.predict(x_test)
# Decision Tree score
print("Decision tree score for test set:")
print("Precision: {:.3f}".format(precision_score(y_test, y_hat_test, average='micro')), "Recall: {:.3f}".format(recall_score(y_test, y_hat_test, average='micro')))
print("F1 score: {:.3f}".format(f1_score(y_test, y_hat_test, average='micro')))
```

```python
#逻辑回归
from sklearn.linear_model import LogisticRegression
# Fitting a logistic regression model with default parameters
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
# Prediction & Evaluation
y_hat_test = logreg.predict(x_test)
# Logistic Regression score
print("Logistic regression score for test set:")
print("Precision: {:.3f}".format(precision_score(y_test, y_hat_test, average='micro')), "Recall: {:.3f}".format(recall_score(y_test, y_hat_test, average='micro')))
print("F1 score: {:.3f}".format(f1_score(y_test, y_hat_test, average='micro')))
```
